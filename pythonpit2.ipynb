```python
import matplotlib.pyplot as plt # for plotting
```

```python
plt.hist(df['latitude'], bins=100)  # Command for creating the histogram
plt.xlabel('Latitude value')  # Command for labeling the x-axis
plt.show()  # Command for displaying the plot
```

```python
#Select just data relevant to Food production

plot_data = df[df['Element'] == 'Food']

# Group the data by the Production Area and sum the values for 2013.

plot_data = plot_data.groupby('Area')['Y2013'].sum() # Sort the values in ascending order, and plot

# the last 10 values in a barchart

plot_data.sort_values()[-10:].plot(kind='bar')

plt.title("Top Ten Food Producers")

plt.ylabel("Food produced (tonnes)")
```
